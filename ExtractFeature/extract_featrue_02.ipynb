```python
import matplotlib.pyplot as plt
%matplotlib inline

class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = {'batch':[], 'epoch':[]}
        self.accuracy = {'batch':[], 'epoch':[]}
        self.val_loss = {'batch':[], 'epoch':[]}
        self.val_acc = {'batch':[], 'epoch':[]}

    def on_batch_end(self, batch, logs={}):
        self.losses['batch'].append(logs.get('loss'))
        self.accuracy['batch'].append(logs.get('acc'))
        self.val_loss['batch'].append(logs.get('val_loss'))
        self.val_acc['batch'].append(logs.get('val_acc'))

    def on_epoch_end(self, batch, logs={}):
        self.losses['epoch'].append(logs.get('loss'))
        self.accuracy['epoch'].append(logs.get('acc'))
        self.val_loss['epoch'].append(logs.get('val_loss'))
        self.val_acc['epoch'].append(logs.get('val_acc'))

    def loss_plot(self, loss_type):
        iters = range(len(self.losses[loss_type]))
        plt.figure()
        # acc
        plt.plot(iters, self.accuracy[loss_type], 'r', label='train acc')
        # loss
        plt.plot(iters, self.losses[loss_type], 'g', label='train loss')
        if loss_type == 'epoch':
            # val_acc
            plt.plot(iters, self.val_acc[loss_type], 'b', label='val acc')
            # val_loss
            plt.plot(iters, self.val_loss[loss_type], 'k', label='val loss')
        plt.grid(True)
        plt.xlabel(loss_type)
        plt.ylabel('acc-loss')
        plt.legend(loc="upper right")
        plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers

kValue_of_KFold = 3
list_model_mention = []
matrix_score_mention = []
matrix_history_mention = []

for i in range(len(list_nodeName)):    
#     model_mention = Sequential()
#     model_mention.add(Dense(units=256, activation='relu', input_dim=len(list_vecOfText[0])))
#     model_mention.add(Dense(units=128, activation='relu'))
#     model_mention.add(Dense(units=64, activation='relu'))
#     model_mention.add(Dense(units=1, activation='sigmoid'))

#     model_mention.compile(loss='binary_crossentropy',
#                   optimizer=optimizers.RMSprop(),
#                   metrics=['accuracy'])

# #     print(model_mention.summary())

#     check_point_mention = keras.callbacks.ModelCheckpoint('./CheckPoint/guangyi3000xianbingshi/guangyi3000xianbingshi_modelMentionmention{}_checkpoint.hdf5'.format(i), monitor='val_loss', verbose=0, 
#                                                   save_best_only=True, save_weights_only=False, mode='auto', period=1)
#     early_stop_mention = keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=30, verbose=0,
#                                   mode='auto', baseline=None, restore_best_weights=False)
#     #创建一个实例history
#     history_mention = LossHistory()
#     callbacks_mention = [check_point_mention, early_stop_mention,history_mention]

#     def classifier(input_shape):
#         """classifier network.
#         Args:
#         input_shape: the desired input shape (e.g.: (latent_space_size))
#         Returns:
#         C: The classifier model
#         """
#         inputs = tf.keras.layers.Input(input_shape)
#         net = tf.keras.layers.Dense(units=256, activation=tf.nn.elu, name="fc1")(inputs)
#         net = tf.keras.layers.Dense(units=128, activation=tf.nn.elu, name="fc2")(net)
#         net = tf.keras.layers.Dense(units=64, activation=tf.nn.elu, name="fc3")(net)
#         net = tf.keras.layers.Dense(units=1, activation=tf.nn.softmax, name="C")(net)
#         C = tf.keras.Model(inputs=inputs, outputs=net)
#         return C

    
    X = np.array(list_vecOfText)
    y = np.array(matrix_label_mention[i])
    
    list_score_mention = []
    list_history_mention = []
    ii = 0
    for train_index, test_index in KFold(kValue_of_KFold).split(X):
        model_mention = Sequential()
        model_mention.add(Dense(units=256, activation='relu', input_dim=len(list_vecOfText[0])))
        model_mention.add(Dense(units=128, activation='relu'))
        model_mention.add(Dense(units=64, activation='relu'))
        model_mention.add(Dense(units=1, activation='sigmoid'))

        model_mention.compile(loss='binary_crossentropy',
                      optimizer=optimizers.RMSprop(),
                      metrics=['accuracy'])

    #     print(model_mention.summary())

        check_point_mention = keras.callbacks.ModelCheckpoint('./CheckPoint/guangyi3000xianbingshi/guangyi3000xianbingshi_modelMentionmention{}_checkpoint.hdf5'.format(i), monitor='val_loss', verbose=0, 
                                                      save_best_only=True, save_weights_only=False, mode='auto', period=1)
        early_stop_mention = keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=30, verbose=0,
                                      mode='auto', baseline=None, restore_best_weights=False)
        #创建一个实例history
        history_mention = LossHistory()
        callbacks_mention = [check_point_mention, early_stop_mention,history_mention]
        
        
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        model_mention.fit(X_train, y_train, epochs=1000, batch_size=32,
                     callbacks=callbacks_mention, validation_split=0.2)

        score_mention = model_mention.evaluate(X_test, y_test, verbose=0)
        print('\n{}.{})Test loss of feature:{} Fold:{} :'.format(i,ii,list_nodeName[i],ii), score_mention[0])
        print('{}.{})Test accuracy of feature:{} Fold:{} :'.format(i,ii,list_nodeName[i], ii), score_mention[1])
        print('\n\n\n')
        list_score_mention.append(score_mention)
        
        #绘制acc-loss曲线
#         history_mention.loss_plot('epoch')
        list_history_mention.append(history_mention)

        # predict_mention = model_mention.predict(np.array(list_vecOfNodeText))
        # print(predict_mention)
        # print(matrix_labelmention)
        ii += 1
    
    list_model_mention.append(model_mention)
    matrix_score_mention.append(list_score_mention)
    matrix_history_mention.append(list_history_mention)
```
